```python
import matplotlib.pyplot as plt
import csv

# Data storage for timSort and powerSort
timSort_times = []
powerSort_times = []
sizes_timSort = []
sizes_powerSort = []

# Read the CSV file and extract data
with open('C:/Users/rvnyamagouda/OneDrive - California State University Chico/Math_608/project_608/sort_times.csv', mode='r') as file:
    reader = csv.reader(file)
    next(reader)  # Skip the header row
    for row in reader:
        timSort_times.append(float(row[0]))
        sizes_timSort.append(int(row[1]))
        powerSort_times.append(float(row[2]))
        sizes_powerSort.append(int(row[3]))

# Plotting the data
plt.figure(figsize=(10, 6))

# Plot for timSort
plt.plot(sizes_timSort, timSort_times, label='timSort', color='b', marker='o')

# Plot for powerSort
plt.plot(sizes_powerSort, powerSort_times, label='powerSort', color='r', marker='x')

# Adding labels and title
plt.xlabel('Array Size')
plt.ylabel('Time Taken (ms)')
plt.title('Sorting Time vs Array Size for timSort and powerSort')
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
